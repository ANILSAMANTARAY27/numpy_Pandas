```python
corrected_dates = []

for i in audible['releasedate']:
    try:
        f_date = pd.to_datetime(i, format='%d-%m-%Y', errors='raise')
    except ValueError:
        try:
            f_date = pd.to_datetime(i, format='%Y-%m-%d', errors='raise')
        except ValueError:
            try:
                f_date = pd.to_datetime(i, format='%d-%m-%y', errors='raise')
            except ValueError:
                f_date = pd.NaT
    corrected_dates.append(f_date)

audible['releasedate'] = corrected_dates
```

```python
def convert_time(time_str):
    parts = time_str.split()
    hours = 0
    minutes = 0
    
    for i in range(len(parts)):
        if parts[i] == 'hrs' or parts[i] ==  'hr':
            hours = int(parts[i - 1])
        elif parts[i] == 'mins':
            minutes = int(parts[i - 1])
    return f'{hours:02}:{minutes:02}:00'
    
audible['time']=audible['time'].apply(convert_time)
```

```python
import re

def extract_ratings(rating_str):
    pattern = r'(\d+\.\d+|\d+) out of 5'
    match = re.search(pattern, rating_str)

    if match:
        return match.group(1) 
    else:
        return None
audible['stars']=audible['stars'].apply(extract_ratings)
```
